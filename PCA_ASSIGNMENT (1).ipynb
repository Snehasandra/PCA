```python
scaler=StandardScaler()
scaler.fit(wine)
```

```python
pca=PCA(n_components=3)
```

```python
pca.fit(wine)
```

```python
pca_values= pca.transform(wine)
pca_values
```

```python
pca_df=pd.DataFrame(pca_values,columns=['PCA1','PCA2','PCA3'])
pca_df
```

```python
#variance
var=pca.explained_variance_ratio_
var
```

```python
plt.plot(var,color='red')
```

```python
#cumulative varience
var=np.cumsum(np.round(var,decimals=3)*100)
var
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('cumulative explained varience')
```

#Hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
linkages=['single','complete','average','ward']
for i ,linkages in enumerate(linkages):
  model=AgglomerativeClustering(n_clusters=4,linkage=linkages)

  labels= model.fit_predict(pca_values)
  pca_df[f'labels_{linkages}']=labels
```

```python
pca_df
```

```python
#create dendrogram
dendrogram=sch.dendrogram(sch.linkage(pca_values,method='single'))
```

```python
pca_df.groupby('labels_single').mean()
```

```python
sns.scatterplot(pca_values)
```

```python
dendrogram=sch.dendrogram(sch.linkage(pca_values,method='complete'))
```

```python
sns.scatterplot(pca_df,x='PCA1',y='PCA2',hue='labels_complete')
```

```python
dendrogram=sch.dendrogram(sch.linkage(pca_values,method='average'))
```

```python
sns.scatterplot(pca_df,x='PCA1',y='PCA2',hue='labels_average')
```

```python
dendrogram=sch.dendrogram(sch.linkage(pca_values,method='ward'))
```

```python
sns.scatterplot(pca_df,x='PCA1',y='PCA2',hue='labels_ward')
```

#K-MEANS

```python
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
kmeans=KMeans(n_clusters=5)
kmeans.fit_transform(pca_values)
labels=kmeans.fit_predict(pca_values)
labels
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=100)
  kmeans.fit_transform(pca_values)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('Number of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=8)
kmeans.fit_transform(pca_values)
labels=kmeans.predict(pca_values)
labels
```

```python
plt.scatter(x=pca_values[:,0],y=pca_values[:,1],c=labels)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```
